```python
import pandas as pd
import os
from pathlib import Path 
```

```python
merged_df = pd.read_csv('data/merged.csv', index_col=0)
```

```python
merged_df.head(5)
```

```python
merged_df.groupby('sector')['ticker'].nunique()
```

```python
# Consumer staples
cs_df = merged_df[merged_df['sector'] == 'Consumer Staples']
cs_df = cs_df.reset_index(drop=True)
cs_df
```

```python
# Number of companies

cs_df['ticker'].unique()
```

```python
# Create a df by ticker
cl = cs_df[cs_df['ticker'] == 'CL']
chd = cs_df[cs_df['ticker'] == 'CHD']
hsy = cs_df[cs_df['ticker'] == 'HSY']
cpb = cs_df[cs_df['ticker'] == 'CPB']
```

```python
cl.to_csv("data/CL.csv", index=False)
chd.to_csv("data/CHD.csv", index=False)
hsy.to_csv("data/HSY.csv", index=False)
cpb.to_csv("data/CPB.csv", index=False)
```
